#### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

import pickle
```

#### Dataset Visualization

```python
# Dataset Path
data_path = r"../data/rural.csv"

# load dataset
df = pd.read_csv(data_path)

# Display the first rows
print(df.head())
```

```python
# dataset info
print(df.info())
```

#### Fixing Missing values

```python
# removing missing values
df = df.dropna()
print("New Shape: ", df.shape)
```

```python
# display features
print(df.columns)
```

```python
# conversion to datetime
df['datetime'] = pd.to_datetime(df['datetime'])

df['hour'] = df['datetime'].dt.hour
df['day'] = df['datetime'].dt.day
df['month'] = df['datetime'].dt.month

df['loss_kwh'] = df['energy_supplied_kwh'] - df['energy_billed_kwh']
df['loss_ratio'] = df['loss_kwh'] / df['energy_supplied_kwh']
df['rolling_loss_mean'] = df['loss_ratio'].rolling(24).mean()
df['rolling_loss_std'] = df['loss_ratio'].rolling(24).std()

df.head()
```

#### Feature Scaling

```python
# Defining the features

features = df [
    [
        'energy_supplied_kwh',
        'energy_billed_kwh',
        'loss_ratio',
        'hour',
        'day',
        'month',
        'rolling_loss_mean',
        'rolling_loss_std'
    ]
]

scaler = StandardScaler()
X_Scaled = scaler.fit_transform(features)
```

```python
# Training the model

model = IsolationForest (
    contamination=0.03,
    n_estimators=200,
    random_state=42
)

df['anomaly'] = model.fit_predict(X_Scaled)
df['anomaly_score'] = model.decision_function(X_Scaled)
```

```python
df.to_csv("../outputs/rural_anomalies.csv", index=False)
```

```python
print(df[['datetime', 'loss_ratio', 'anomaly', 'anomaly_score', 'rolling_loss_mean', 'rolling_loss_std']].head())
```

```python
# Check the number of anomalies
print(df['anomaly'].value_counts())
```

#### Visualize Anomalies

```python
anomalies = df[df['anomaly'] == -1]
plt.figure(figsize=(14,6))

# Normal data
plt.plot (
    df['datetime'],
    df['loss_ratio'],
    label="Loss Ratio",
    color='blue',
    alpha=0.5
)

plt.scatter (
    anomalies['datetime'],
    anomalies['loss_ratio'],
    color='red',
    label='Anomaly',
    s=30,
    zorder=5
)

plt.title("Transformer Energy Loss Anomaly Detection")
plt.xlabel("Datetime")
plt.ylabel("Loss Ratio")

plt.legend()
plt.show()
```

#### Saving the ML model

```python
import os
os.makedirs('../models', exist_ok=True)

with open('../models/voltguard_rural_model.pkl', 'wb') as f:
    pickle.dump({'model': model, 'scaler': scaler}, f)

print("Model saved successfully.")
```

#### Exporting the Anomaly results

```python
os.makedirs('../outputs', exist_ok=True)

anomalies.to_csv('../outputs/rural_anomalies.csv', index=False)
print("Anomalies exported.")
```
